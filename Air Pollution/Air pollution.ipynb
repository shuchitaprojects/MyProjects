```python
import pandas as pd
import requests

url = 'http://api.openweathermap.org/data/2.5/air_pollution/forecast?lat=26.45&lon=80.33&appid=5edc92589172b7942928857b49aa27c9'


response = requests.get(url)
json_data = response.json()['list']
json_data
```

```python
df = pd.DataFrame.from_records(json_data)
df.head()
```

```python
df.components
```

```python
keys = set()
for d in df['components']:
    keys.update(d.keys())

# Create new columns for each key
for key in keys:
    df[key] = df['components'].apply(lambda x: x.get(key))

# Drop the 'data' column
df = df.drop(columns=['components'])
```

```python
keys = set()
for d in df['main']:
    keys.update(d.keys())

# Create new columns for each key
for key in keys:
    df[key] = df['main'].apply(lambda x: x.get(key))

# Drop the 'data' column
df = df.drop(columns=['main'])
```

```python
df = df.drop(['dt'], axis = 1)
```
