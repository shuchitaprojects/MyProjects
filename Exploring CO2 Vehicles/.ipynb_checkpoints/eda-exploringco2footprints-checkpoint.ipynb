# Import Essential Libraries

```python
# import essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
%matplotlib inline
```

# Load the Dataset

```python
# load the dataset
data = pd.read_csv('/kaggle/input/co2-emissions/CO2 Emissions.csv')

# check the dimension of the dataset
print('Dimension of the dataset: ', data.shape)

# check the attributes in the dataset
print('Attributes in the dataset: ', data.columns.values)

# view the first 5 rows of the dataset
data.head()
```

#### Attributes Description:

1. Make - vehicle manufacturer
2. Model - vehicle model
3. Vehicle class - different vehicle classes, such as sedans, SUVs, trucks, or sports cars
4. Engine Size (L) - size of engine
5. Cylinders - engine cylinder
6. Transmission - type of transmissison (automatic or manual transmission)
7. Fuel Type - Ttpe of fuel used by a vehicle
8. Fuel Consumption City (L/100km) - fuel consumption of a vehicle in city
9. Fuel Consumption Hwy (L/100 km) - fuel consumption of a vehicle in highway
10. Fuel Consumption Comb (L/100 km) - fuel consumption of a vehicle in city and highway (combined) in unit L/km
11. Fuel Consumption Comb (mpg) - fuel consumption of a vehicle in city and highway (combined) in unit miles per gallon
12. CO2 Emissions(g/km) - amount of CO2 gas emitted

# Data Cleaning

```python
# insert a new column 'Transmission_Class' - ['Automatic', 'Manual']
data['Transmission_Class'] = data['Transmission'].apply(lambda x: 'Automatic' if x.startswith('A') else 'Manual')
```

```python
# replace the value in column 'Fuel Type'
data['Fuel Type'] = data['Fuel Type'].replace(['X', 'Z', 'E', 'D', 'N'], ['Regular Gasoline','Premium Gasoline','Ethanol','Diesel','Natural Gas'])
```

```python
# top level overview of the dataset
data.info()
```

# Exploratory Data Analysis
#### Question 1 | What are the top vehicle manufacturers that collectively make up 50% of the vehicles in the dataset, and how does their distribution compare?

```python
manufacturer = data['Make'].value_counts()
val = manufacturer.values
fifty_percent = data.shape[0] / 2

total = 0
count = 0
for i in val:
    total = total + i
    count = count + 1
    if total > fifty_percent:
        break
count

plt.figure(figsize=(15,6))
plt.bar(x=manufacturer.index[:count], height=manufacturer.values[:count], color='indigo')

for index, values in enumerate(manufacturer[:count]):
    plt.text(index, values+8, str(values), ha='center')

plt.grid(alpha=0.4)
plt.title('Top Vehicle Manufacturers', fontsize=15)
plt.show()
```

#### Question 2 | Among the top vehicle manufacturers that collectively represent 50% of the dataset, which ones have the highest and lowest average fuel efficiency, measured by Fuel Consumption Comb (L/100km)?

```python
topManufacturer = manufacturer.index[:count]
fil_data = data[data['Make'].isin(topManufacturer)]

plt.figure(figsize=(15,5))
sns.boxplot(data=fil_data, x=fil_data['Make'], y=fil_data['Fuel Consumption Comb (L/100 km)'], palette='magma')
plt.xlabel('Vehicle Manufacturer')
plt.ylabel('Fuel Efficiency')
plt.grid(alpha=0.4)
plt.title('Comparison of the Vehicle Fuel Efficiency among the Top Manufacturers', fontsize=15)
plt.show()
```

#### Question 3 | Among the top manufacturers, collectively constituting 50% of the dataset, which ones have the highest and lowest average CO2 emissions (g/km)?

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=fil_data, x=fil_data['Make'], y=fil_data['CO2 Emissions(g/km)'], palette='magma')
plt.xlabel('Vehicle Manufacturer')
plt.ylabel('CO2 Emissions (g/km)')
plt.grid(alpha=0.4)
plt.title('Comparison of the Vehicle CO2 Emissions (g/km) among the Top Manufacturers', fontsize=15)
plt.show()
```

#### Question 4 | What is the distribution of vehicle classes in the dataset?

```python
vehicles = data['Vehicle Class'].value_counts().sort_values()

# visualization
plt.figure(figsize=(15,8))
plt.barh(y=vehicles.index, width=vehicles.values, color='indigo')
plt.xlabel('Frequency')

for index, values in enumerate(vehicles):
    plt.text(values+5, index, str(values), va='center')

plt.grid(alpha=0.4)
plt.title('Distribution of Vehicle Classes', fontsize=15)
plt.show()
```

#### Question 5 | Which vehicle type have the highest and lowest average fuel efficiency, measured by Fuel Consumption Comb (L/100km)?

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=data, x=data['Vehicle Class'], y=data['Fuel Consumption Comb (L/100 km)'], palette='magma')
plt.xlabel('Vehicle Type')
plt.xticks(rotation=90)
plt.ylabel('Fuel Efficiency')
plt.grid(alpha=0.4)
plt.title('Comparison of the Fuel Efficiency among Various Vehicle Types', fontsize=15)
plt.show()
```

#### Question 6 | Which vehicle type have the highest and lowest average CO2 emissions (g/km)?

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=data, x=data['Vehicle Class'], y=data['CO2 Emissions(g/km)'], palette='magma')
plt.xlabel('Vehicle Type')
plt.xticks(rotation=90)
plt.ylabel('CO2 Emissions (g/km)')
plt.grid(alpha=0.4)
plt.title('Comparison of the CO2 Emissions (g/km) among tVarious Vehicle Types', fontsize=15)
plt.show()
```

#### Question 7  | Which transmission type (automatic or manual) is more common among different vehicle classes?

```python
plt.figure(figsize=(15,6))
ax=sns.countplot(data=data, x=data['Vehicle Class'], hue=data['Transmission_Class'], palette='magma')
plt.xticks(rotation=90)
plt.xlabel(' ')
plt.ylabel('Frequency')
plt.grid(alpha=0.4)
plt.title('Transmission Type in Different Vehicle Classes', fontsize=15)
plt.show()
```

#### Question 8 | Is there any difference between the fuel efficiency and CO2 emission between automatic and manual transmission vehicles?

```python
figure, axes = plt.subplots(1,2, figsize=(15,5), gridspec_kw={'width_ratios':[1.5,1]})
sns.histplot(data=data, x=data['CO2 Emissions(g/km)'], hue=data['Transmission_Class'], palette='magma', ax=axes[0])
axes[0].grid(alpha=0.4)
sns.violinplot(data=data, x=data['Transmission_Class'], y=data['CO2 Emissions(g/km)'], palette='magma', ax=axes[1])
axes[1].grid(alpha=0.4)
figure.suptitle('CO2 Emission (g/km) between Automatic and Manual Transmission Vehicles', fontsize=15)
plt.tight_layout(pad=1)
plt.show()

figure, axes = plt.subplots(1,2, figsize=(15,5), gridspec_kw={'width_ratios':[1.5,1]})
sns.histplot(data=data, x=data['Fuel Consumption Comb (L/100 km)'], hue=data['Transmission_Class'], palette='magma', ax=axes[0])
axes[0].grid(alpha=0.4)
sns.violinplot(data=data, x=data['Transmission_Class'], y=data['Fuel Consumption Comb (L/100 km)'], palette='magma', ax=axes[1])
axes[1].grid(alpha=0.4)
figure.suptitle('Fuel Efficiency between Automatic and Manual Transmission Vehicles', fontsize=15)
plt.tight_layout(pad=1)
plt.show()
```

#### Question 9 | What is the most common fuel type used by vehicles in the dataset?

```python
fuel_type = data['Fuel Type'].value_counts().sort_values()

plt.figure(figsize=(12,5))
plt.barh(y=fuel_type.index, width=fuel_type.values, color='indigo')
plt.xlabel('Frequency')

for index, values in enumerate(fuel_type):
    plt.text(values+20, index, str(values), va='center')

plt.grid(alpha=0.4)
plt.title('Most Common Fuel Type Used by Vehicles', fontsize=15)
plt.show()
```

#### Question 10 | Is there any difference between the fuel efficiency and CO2 emission between vehicles with different fuel types?

```python
fil_data = data[data['Fuel Type']!='Natural Gas']
labels = ['Regular Gasoline','Premium Gasoline','Ethanol','Diesel']

figure, axes = plt.subplots(2,1, figsize=(10,10))
sns.boxplot(data=fil_data, x=fil_data['Fuel Type'], y=fil_data['CO2 Emissions(g/km)'], palette='magma', ax=axes[0])
axes[0].grid(alpha=0.4)
axes[0].set_xticklabels(labels)
axes[0].set_title('CO2 Emission (g/km)', fontsize=12)
sns.boxplot(data=fil_data, x=fil_data['Fuel Type'], y=fil_data['Fuel Consumption Comb (L/100 km)'], palette='magma', ax=axes[1])
axes[1].grid(alpha=0.4)
axes[1].set_xticklabels(labels[:4])
axes[1].set_title('Fuel Consumption (L/100km)', fontsize=12)
figure.suptitle('Comparison between Different Fuel Type Vehicles', fontsize=15)
plt.tight_layout(pad=1)
plt.show()
```

#### Question 11  | Is there a correlation between the engine size and fuel efficiency (Fuel Consumption Comb)?

```python
plt.figure(figsize=(15,5))
ax=sns.relplot(data=data, x=data['Engine Size(L)'], y=data['Fuel Consumption Comb (L/100 km)'], 
               hue=data['Transmission_Class'], palette='magma', height=6, aspect=1.5)
plt.title('Engine Size vs Fuel Efficiency (in City & Highway)', fontsize=15)
plt.show()
```

#### Question 12 | Is there a discernible relationship between CO2 emissions and fuel consumption (city, highway, or combined)? Additionally, does this relationship differ between vehicles with automatic and manual transmissions?

```python
plt.figure(figsize=(15,5))
sns.relplot(data=data, x=data['CO2 Emissions(g/km)'], y=data['Fuel Consumption City (L/100 km)'], 
               hue=data['Transmission_Class'], palette='magma', height=6, aspect=1.5)
plt.title('CO2 Emissions (g/km) vs Fuel Efficiency (in City) in Automatic & Manual Transmission Vehicles', fontsize=15)
plt.show()

plt.figure(figsize=(15,5))
sns.relplot(data=data, x=data['CO2 Emissions(g/km)'], y=data['Fuel Consumption Hwy (L/100 km)'], 
               hue=data['Transmission_Class'], palette='magma', height=6, aspect=1.5)
plt.title('CO2 Emissions (g/km) vs Fuel Efficiency (in Highway) in Automatic & Manual Transmission Vehicles', fontsize=15)
plt.show()

plt.figure(figsize=(15,5))
sns.relplot(data=data, x=data['CO2 Emissions(g/km)'], y=data['Fuel Consumption Comb (L/100 km)'], 
               hue=data['Transmission_Class'], palette='magma', height=6, aspect=1.5)
plt.title('CO2 Emissions (g/km) vs Fuel Efficiency (in city & Highway) in Automatic & Manual Transmission Vehicles', fontsize=15)
plt.show()
```

#### Question 13 | Does the relationship between between CO2 emissions and fuel consumption (city, highway, or combined) varies between vehicles with different fuel type?

```python
plt.figure(figsize=(15,5))
ax=sns.relplot(data=data, x=data['CO2 Emissions(g/km)'], y=data['Fuel Consumption City (L/100 km)'], 
               hue=data['Fuel Type'], palette='magma', height=6, aspect=1.5)
plt.title('CO2 Emissions (g/km) vs Fuel Efficiency (in City) in Various Fuel Type Vehicles', fontsize=15)
plt.show()

plt.figure(figsize=(15,5))
ax=sns.relplot(data=data, x=data['CO2 Emissions(g/km)'], y=data['Fuel Consumption Hwy (L/100 km)'], 
               hue=data['Fuel Type'], palette='magma', height=6, aspect=1.5)
plt.title('CO2 Emissions (g/km) vs Fuel Efficiency (in Highway) in Various Fuel Type Vehicles', fontsize=15)
plt.show()

plt.figure(figsize=(15,5))
ax=sns.relplot(data=data, x=data['CO2 Emissions(g/km)'], y=data['Fuel Consumption Comb (L/100 km)'], 
               hue=data['Fuel Type'], palette='magma', height=6, aspect=1.5)
plt.title('CO2 Emissions (g/km) vs Fuel Efficiency (in City & Highway) in Various Fuel Type Vehicles', fontsize=15)
plt.show()
```

#### Correlation between engine size, cylinders, fuel consumption (in city, in highway, in city and highway) and the CO2 emission.

```python
# compute the Pearson's correlation between features
num_vars = ['Engine Size(L)', 'Cylinders','Fuel Consumption City (L/100 km)','Fuel Consumption Hwy (L/100 km)', 'Fuel Consumption Comb (L/100 km)', 'Fuel Consumption Comb (mpg)', 'CO2 Emissions(g/km)']
correlation = data[num_vars].corr()

# visualize the correlation in heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlation, annot=True, cmap='plasma')
plt.show()
```
